```python
df = pd.read_csv('fashion-mnist_train.csv')
df.head(10)
```

```python
df.describe()
```

```python
target = np.array([df['label']]).T
target.shape
```

```python
features = np.array(df.iloc[:,1:])
features.shape
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
y = to_categorical(target) 
y.shape
```

```python
X_train,X_val,y_train,y_val = train_test_split(features,y,test_size=0.2,random_state=0)
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train,test_size=0.21,random_state=0)
```

```python
img_row ,img_col = 28,28

X_train = X_train.reshape(X_train.shape[0],img_row,img_col,1)
X_test = X_test.reshape(X_test.shape[0],img_row,img_col,1)
X_val = X_val.reshape(X_val.shape[0],img_row,img_col,1)

X_train = X_train.astype('float32')
X_val = X_val.astype('float32')
X_test = X_test.astype('float32')

X_train/=255
X_test/=255
X_val/=255
```

```python
X_train.shape
```

```python
from keras import layers
from keras import models
```

```python
model = models.Sequential()
model.add(layers.Conv2D(28, (3, 3), activation='relu',
                        input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(56, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(102, (3, 3), activation='relu'))
# model.add(layers.MaxPooling2D((2, 2)))
# model.add(layers.Conv2D(102, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(784, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
model.fit(X_train,y_train,epochs=3,batch_size=32,validation_data=(X_val,y_val))
```

```python
model.evaluate(X_test,y_test)
```
